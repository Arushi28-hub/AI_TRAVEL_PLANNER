```python
import pandas as pd

df = pd.read_csv("processed_travel_dataset.csv")
```

```python
df.head()
```

```python
def attractions_per_day(days):

    return days * 3
```

```python
attractions_per_day(2)
```

```python
def estimate_trip_cost(
    entrance_fee,
    days
):

    food_cost = days * 400

    transport_cost = days * 300

    total_cost = (
        entrance_fee
        +
        food_cost
        +
        transport_cost
    )

    return total_cost
```

```python
estimate_trip_cost(
    entrance_fee=500,
    days=2
)
```

```python
def plan_trip(
    interest,
    budget,
    days
):

    num_places = attractions_per_day(days)

    matches = df[
        df["Tags"].str.contains(
            interest,
            case=False,
            na=False
        )
    ]

    matches = matches.sort_values(
        by="Student Score",
        ascending=False
    )

    selected_places = []

    total_fee = 0

    for _, row in matches.iterrows():

        fee = row["Entrance Fee in INR"]

        estimated_total = estimate_trip_cost(
            total_fee + fee,
            days
        )

        if estimated_total <= budget:

            selected_places.append(row)

            total_fee += fee

        if len(selected_places) >= num_places:

            break

    result = pd.DataFrame(
        selected_places
    )

    return result
```

```python
trip = plan_trip(
    interest="Historical",
    budget=5000,
    days=2
)

trip[
    [
        "Name",
        "City",
        "Entrance Fee in INR",
        "Student Score"
    ]
]
```

```python
total_entry_fee = trip[
    "Entrance Fee in INR"
].sum()

final_cost = estimate_trip_cost(
    total_entry_fee,
    days=2
)

print(final_cost)
```

```python
budget = 5000

remaining = budget - final_cost

print(remaining)
```

```python
def plan_trip(
    interest,
    budget,
    days,
    city=None
):

    num_places = attractions_per_day(days)

    matches = df[
        df["Tags"].str.contains(
            interest,
            case=False,
            na=False
        )
    ]

    if city:
        matches = matches[
            matches["City"].str.contains(
                city,
                case=False,
                na=False
            )
        ]

    matches = matches.sort_values(
        by="Student Score",
        ascending=False
    )

    selected_places = []

    total_fee = 0

    for _, row in matches.iterrows():

        fee = row["Entrance Fee in INR"]

        estimated_total = estimate_trip_cost(
            total_fee + fee,
            days
        )

        if estimated_total <= budget:

            selected_places.append(row)

            total_fee += fee

        if len(selected_places) >= num_places:
            break

    return pd.DataFrame(selected_places)
```

```python
trip = plan_trip(
    interest="Historical",
    budget=5000,
    days=2,
    city="Delhi"
)

trip[["Name", "City", "Entrance Fee in INR"]]
```

```python
def trip_summary(
    trip,
    budget,
    days
):

    total_entry_fee = trip[
        "Entrance Fee in INR"
    ].sum()

    total_cost = estimate_trip_cost(
        total_entry_fee,
        days
    )

    remaining_budget = (
        budget - total_cost
    )

    print("===== TRIP SUMMARY =====")

    print(
        f"Attractions Selected: {len(trip)}"
    )

    print(
        f"Entry Fees: ₹{total_entry_fee}"
    )

    print(
        f"Estimated Total Cost: ₹{total_cost}"
    )

    print(
        f"Remaining Budget: ₹{remaining_budget}"
    )

    return total_cost
```

```python
trip_summary(
    trip,
    budget=5000,
    days=2
)
```
