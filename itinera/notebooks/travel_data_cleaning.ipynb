```python
import pandas as pd
import numpy as np
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df["Weekly Off"] = df["Weekly Off"].fillna("Open All Days")
```

```python
df["Establishment Year"].unique()[:30]
```

```python
import numpy as np

df["Establishment Year"] = df["Establishment Year"].replace(
    "Unknown",
    np.nan
)
```

```python
df["Establishment Year"].isnull().sum()
```

```python
df["Establishment Year"].notnull().sum()
```

```python
df["Establishment Year"] = pd.to_numeric(
    df["Establishment Year"],
    errors="coerce"
)
```

```python
CURRENT_YEAR = 2026

df["Attraction Age"] = (
    CURRENT_YEAR - df["Establishment Year"]
)
```

```python
median_age = df["Attraction Age"].median()

df["Attraction Age"] = df["Attraction Age"].fillna(
    median_age
)
```

```python
df["Year Known"] = np.where(
    df["Establishment Year"].isnull(),
    "Unknown",
    "Known"
)
```

```python
def age_category(age):

    if age < 50:
        return "Modern"

    elif age < 200:
        return "Historic"

    elif age < 500:
        return "Heritage"

    else:
        return "Ancient"
```

```python
df["Age Category"] = df["Attraction Age"].apply(
    age_category
)
```

```python
df["Year Known"] = np.where(
    df["Establishment Year"].isnull(),
    "Unknown",
    "Known"
)
```

```python
df["Type"].value_counts()
```

```python
df["Best Time to visit"].value_counts()
```

```python
for x in df["Best Time to visit"].unique():
    print(repr(x))
```

```python
df["Best Time to visit"] = (
    df["Best Time to visit"]
    .str.strip()
)
```

```python
df["Best Time to visit"] = (
    df["Best Time to visit"]
    .replace("Anytime", "All")
)
```

```python
df["Best Time to visit"].value_counts()
```

```python
df.head()
```

```python
df.columns
```

```python
def budget_category(fee):

    if fee <= 50:
        return "Very Low"

    elif fee <= 200:
        return "Low"

    elif fee <= 500:
        return "Medium"

    else:
        return "High"

df["Budget Category"] = df["Entrance Fee in INR"].apply(
    budget_category
)
```

```python
df[["Entrance Fee in INR","Budget Category"]].head()
```

```python
def duration_category(hours):

    if hours <= 2:
        return "Quick Visit"

    elif hours <= 4:
        return "Half Day"

    else:
        return "Full Day"

df["Duration Category"] = df[
    "time needed to visit in hrs"
].apply(duration_category)
```

```python
df[
    ["time needed to visit in hrs",
     "Duration Category"]
].head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df["rating_scaled"] = scaler.fit_transform(
    df[["Google review rating"]]
)
```

```python
df["reviews_scaled"] = scaler.fit_transform(
    df[["Number of google review in lakhs"]]
)
```

```python
df["fee_scaled"] = scaler.fit_transform(
    df[["Entrance Fee in INR"]]
)
```

```python
df["Student Score"] = (
    0.5 * df["rating_scaled"]
    +
    0.3 * df["reviews_scaled"]
    +
    0.2 * (1 - df["fee_scaled"])
)
```

```python
df[
    ["Name","Student Score"]
].sort_values(
    by="Student Score",
    ascending=False
).head()
```

```python
import numpy as np

df["Popularity Score"] = (
    df["Google review rating"]
    *
    np.log1p(
        df["Number of google review in lakhs"]
    )
)
```

```python
df["Tags"] = (
    df["Type"].astype(str)
    + " "
    + df["Significance"].astype(str)
    + " "
    + df["Budget Category"].astype(str)
    + " "
    + df["Duration Category"].astype(str)
    + " "
    + df["Best Time to visit"].astype(str)
)
```
