```python
import os
from dotenv import load_dotenv
from google import genai
load_dotenv()
```

```python
client = genai.Client(
    api_key=os.getenv("GEMINI_API_KEY")
)
```

```python
def create_prompt(
    budget,
    days,
    interest,
    city,
    attractions
):

    return f"""
Create a detailed {days}-day travel itinerary for a student.

Budget: ₹{budget}

Interest: {interest}

City: {city}

Recommended Attractions:
{', '.join(attractions)}

Include:

1. Day-wise itinerary

2. Estimated budget breakdown

3. Food suggestions

4. Transport suggestions

5. Student travel tips

Make it realistic and budget-friendly.
"""
```

```python
attractions = [
    "India Gate",
    "Red Fort",
    "Qutub Minar"
]

prompt = create_prompt(
    budget=5000,
    days=2,
    interest="Historical",
    city="Delhi",
    attractions=attractions
)

print(prompt)
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```
